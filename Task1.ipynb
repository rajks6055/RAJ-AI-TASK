   <h1 style="color:green;">AGE DISTRIBUTION</h1>

```python
age_counts = df['age'].value_counts().sort_index()
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart
sns.countplot(data=df, x='age', order=sorted(df['age'].unique()), ax=axs[0])
axs[0].set_title('Age Distribution (Bar Chart)')
axs[0].tick_params(axis='x', rotation=45)

# Pie chart
axs[1].pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=0)
axs[1].set_title('Age Distribution (Pie Chart)')
axs[1].axis('equal')

plt.tight_layout()
plt.show()

```

<h5 style="color :blue">From Graph we can see that Age Grop 70_80 has higher no of people(25.6%)</h5>

<h1 style="color:green">Gender Distribution Across Age Groups</h1>

```python

fig, ax = plt.subplots(figsize=(10, 6))  


sns.countplot(data=df, x='age', hue='gender', ax=ax,
              order=sorted(df['age'].unique()),  
              hue_order=sorted(df['gender'].unique())) 

ax.set_title('Gender Distribution Across Age Groups')
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')
ax.tick_params(axis='x', rotation=45) 
ax.legend(title='Gender')


plt.tight_layout()
plt.show()

```

<h5 style="color:Blue">Female are larger in number as compared to Male</h5>

<h1 style="color:green">Lab Procedures Across Age Groups</h1>

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x='age', y='num_lab_procedures', order=sorted(df['age'].unique()))
plt.title('Lab Procedures Across Age Groups')
plt.xticks(rotation=45)
plt.show()
```

<h1 style="color:green">Top 10 Most Prescribed Medications</h1>

```python
med_cols = ['metformin', 'repaglinide', 'nateglinide','chlorpropamide', 'glimepiride','acetohexamide', 'glipizide',
            'glyburide','tolbutamide', 'pioglitazone', 'rosiglitazone','acarbose','miglitol','troglitazone','tolazamide',
            'examide','citoglipton' ,'insulin','glyburide-metformin','glipizide-metformin','glimepiride-pioglitazone','metformin-rosiglitazone','metformin-pioglitazone']


med_counts = {col: (df[col] != 'No').sum() for col in med_cols}
top_meds = pd.DataFrame(list(med_counts.items()), columns=['Medication', 'Count'])
total_prescribed = top_meds['Count'].sum()
top_meds['Percentage'] = (top_meds['Count'] / total_prescribed) * 100
top_meds = top_meds.sort_values(by='Count', ascending=False)
print(top_meds)
```

```python
top_10_meds = top_meds.head(10)
plt.figure(figsize=(12, 5))
sns.barplot(data=top_10_meds, x='Percentage', y='Medication')

plt.title('Top 10 Most Prescribed Medications (Percentage)')
plt.xlabel('Percentage (%)')
plt.ylabel('Medication')

plt.tight_layout()
plt.show()
```

<h5 style="color:blure"> Insulin is most used presciption</h5>

<h1 style="color :green">Readmission Status</h1>

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))


sns.countplot(data=df, x='readmitted', ax=axs[0])
axs[0].set_title('Readmission Distribution')


sns.boxplot(data=df, x='readmitted', y='time_in_hospital', ax=axs[1])
axs[1].set_title('Time in Hospital by Readmission Status')
axs[1].set_ylabel('Time_in_hospital_(No. of Days)') 

plt.tight_layout()
plt.show()


```

<h5 style="color:blue">we can clearly say that most of patient are not readmitted and spent very less time in hospital
</h5>

<h1 style="color :green">Time Vs Age Grp</h1>

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='age', y='time_in_hospital')
plt.title('Time in Hospital by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Days in Hospital')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h2 style="color:GREEN">Admission Based on outpatient,inpatient,emergency</h2>

```python

outpat_count = df['number_outpatient'].sum()
inpat_count = df['number_inpatient'].sum()
emerg_count = df['number_emergency'].sum()


labels = ['Outpatient', 'Inpatient', 'Emergency']
sizes = [outpat_count, inpat_count, emerg_count]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Patient Distribution by Admission Type')
plt.axis('equal')
plt.show()
```
